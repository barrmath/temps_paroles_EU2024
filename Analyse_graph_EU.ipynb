Cette liste est conçu avec une règle : chaque liste doit avoir le même temps de parole. Ainsi une chaîne qui lutte contre les autres en mettant en avant une liste peut représentative va donc être classé non-égalitaire.<br>

Vue générale :

```python
# preparation data
eu2024_couleur_gen = eu2024.groupby(by='Couleur').agg({'Durée en seconde':'sum'}).reset_index()
eu2024_couleur_gen["Temps de parole en heure"] = eu2024_couleur_gen['Durée en seconde']/3600
eu2024_couleur_gen = eu2024_couleur_gen.sort_values(by='Temps de parole en heure',ascending=False)

# couleurs
# https://fr.wikipedia.org/wiki/Mod%C3%A8le:Infobox_Parti_politique_fran%C3%A7ais/couleurs
colors = ["#0D378A", "#ffeb00","#bb1840","#ff8080","#0066cc","#00c000","#FF00FF","#eeeeee"]
sns.set_palette(sns.color_palette(colors))

# figure 
fig,ax =plt.subplots()
sns.barplot(data=eu2024_couleur_gen,x='Couleur',y='Temps de parole en heure',hue='Couleur')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("Etiquette politique")
ax.set_title("Temps de parole tout média confondu")
plt.show()
```

Alors il faut aussi prendre en compte l'audimat des chaînes. En 2022 les chaînes de télévision TF1, France2 et France 3 representent 30 % de l'audimat. Elles ont donc une plus grande écoute que les autres chaînes.<br>

```python
# preparation data
eu2024_sort = eu2024.sort_values(by="Temps total",ascending=False)

# couleurs
# https://fr.wikipedia.org/wiki/Mod%C3%A8le:Infobox_Parti_politique_fran%C3%A7ais/couleurs

# figure 
fig,ax =plt.subplots()
fig.set_size_inches(20,10)
sns.barplot(data=eu2024_sort,x='Nom du media',y='Taux temps de parole',hue='Couleur',palette=colors,errorbar=None,estimator='sum')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("Etiquette politique")
ax.set_title("Temps de parole tout media confondu")
plt.show()
```

Si on regarde les pourcentages consacrés aux extrêmes droites, on s'aperçoit que leur temps de parole correspond à environ 30 % du temps d'antenne.<br>
BFM business ne s'intéresse qu'au discours d'état. 100 % de discours


```python
#voir si un groupe de media favorise une couleur 
data = eu2024.groupby(by=['proprio du media','Couleur']).agg({'Durée en seconde':'sum'}).reset_index()
# je dois refaire un pourcentage de temps par rapport au proprio du media
data_proprio = data.groupby(by='proprio du media').agg({'Durée en seconde':'sum'}).reset_index().rename(columns={'Durée en seconde':'Temps total'})
data = pd.merge(data,data_proprio,on='proprio du media')
data['Taux temps de parole'] = data['Durée en seconde']/data['Temps total']*100
# figure 
fig,ax =plt.subplots()
fig.set_size_inches(20,10)
sns.barplot(data=data,x='proprio du media',y='Taux temps de parole',hue='Couleur',palette=colors,errorbar=None,estimator='sum')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("propriétaire des medias")
ax.set_title("Taux de temps de parole par propriétaire de media")
plt.show()
```

Ce graphe permets de se faire une idée de la couleur politique des chaine apartenant à un groupe.<br>
La colonne "macron parle" permet de se faire aussi une idée de l'interet pour la politique electorale pour ce groupe. Au plus la colonne est haute, au moins le groupe diffuse de contenu politique.<br>


```python
#voir si un groupe de media favorise une couleur 
data = eu2024.groupby(by=['Listes candidates']).agg({'Durée en seconde':'sum','resultat2024':'first'}).reset_index()
# je dois refaire un pourcentage de temps par rapport au proprio du media
data['Temps total'] = data['Durée en seconde'].sum()
data['Taux temps de parole'] = data['Durée en seconde']/data['Temps total']*100
# figure 
fig,ax =plt.subplots()
sns.scatterplot(data=data,x='Taux temps de parole',y='resultat2024')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("Taux temps de paroles")
ax.set_title("Taux de temps de parole et resultats")
plt.show()
```
