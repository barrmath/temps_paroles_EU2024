Vue générale :

```python
# preparation data
eu2024_couleur_gen = eu2024.groupby(by='Couleur').agg({'Durée en seconde':'sum'}).reset_index()
eu2024_couleur_gen["Temps de parole en heure"] = eu2024_couleur_gen['Durée en seconde']/3600
eu2024_couleur_gen = eu2024_couleur_gen.sort_values(by='Temps de parole en heure',ascending=False)

# couleurs
# https://fr.wikipedia.org/wiki/Mod%C3%A8le:Infobox_Parti_politique_fran%C3%A7ais/couleurs
colors = ["#0D378A", "#ffeb00","#bb1840","#ff8080","#0066cc","#00c000","#FF00FF","#eeeeee"]
sns.set_palette(sns.color_palette(colors))

# figure 
fig,ax =plt.subplots()
sns.barplot(data=eu2024_couleur_gen,x='Couleur',y='Temps de parole en heure',hue='Couleur')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("Etiquette politique")
ax.set_title("Temps de parole tout média confondu")
plt.show()
```

Selon l'indice de Gini, et les graphiques, le temps de paroles n'est pas égalitaire.<br>
En effet, pour assurer le pluralisme politique, l'ARCOM applique une règle d'équité du temps de paroles.<br>
Les temps de paroles sont donc surveillés en fonction de :
<ul>
    <li>Des résultats des élections précedentes</li>
    <li>Des sondages</li>
    <li>De l'actualité</li>
</ul>

```python
#voir si un groupe de media favorise une couleur 

data = eu2024.groupby(by=['proprio du media','Couleur']).agg({'Durée en seconde':'sum'}).reset_index()

# je dois refaire un pourcentage de temps par rapport au proprio du media

data_proprio = data.groupby(by='proprio du media').agg({'Durée en seconde':'sum'}).reset_index().rename(columns={'Durée en seconde':'Temps total'})
data = pd.merge(data,data_proprio,on='proprio du media')
data['Taux temps de parole'] = data['Durée en seconde']/data['Temps total']*100

# figure 

fig,ax =plt.subplots()
fig.set_size_inches(15,10)
sns.barplot(data=data,x='proprio du media',y='Taux temps de parole',hue='Couleur',palette=colors,errorbar=None,estimator='sum')
ax.tick_params(axis='x', labelrotation=90)
ax.set_xlabel("Propriétaire des medias")
ax.set_title("Taux de temps de parole par propriétaire de media")
plt.show()
```

Ce graphe permet de se faire une idée de la couleur politique des chaînes appartenant à un groupe.<br>


```python
#voir si un groupe de media favorise une couleur 

data = eu2024.groupby(by=['Listes candidates']).agg({'Durée en seconde':'sum','resultat2024':'first'}).reset_index()

# je dois refaire un pourcentage de temps par rapport au proprio du media

data['Temps total'] = data['Durée en seconde'].sum()
data['Taux temps de parole'] = data['Durée en seconde']/data['Temps total']*100

# figure 

fig,ax =plt.subplots()
sns.scatterplot(data=data,x='Taux temps de parole',y='resultat2024')
ax.tick_params(axis='x', labelrotation=90)
ax.set_ylabel("Résultat à l'élection")
ax.set_xlabel("Taux temps de parole")
ax.set_title("Taux de temps de parole et resultats")
plt.show()
```

```python
# test de correlation 

print("Le taux de corrélation de Pearson est de :",round(data['Taux temps de parole'].corr(data['resultat2024']),2))
```

Il y a donc une forte corrélation entre le taux de temps de parole et les résultats à l'élection.<br>


Pour résumer :<br>
Il n'y a pas d'égalité de temps de parole sauf pendant une période de la campagne présidentielle.<br>
Les medias sont contrôlé a posteriori, par l'ARCOM.<br>
Les medias sont soumis à une équité du temps de parole.<br>
Les temps de parole sont influencés par le résultat des précédentes élections, les sondages et aussi l'actualité.<br>
Le temps de parole est corrélé aux résultats des élections.<br>
<br>
Donc il y a un intérêt pour les parties d'avoir un maximum de temps de parole. Les sondages sont un moyen d'y arriver.<br>
Mais l'actu peut aussi influencer le temps de parole. Ainsi, une phrase choc qui fait parler peux fournir du temps de parole.<br>
Ainsi la possession des médias + institut de sondage permettrai d'avoir une très forte influence en matière de politique du pays.<br>
